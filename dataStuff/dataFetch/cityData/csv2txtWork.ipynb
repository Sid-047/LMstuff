```python
import pandas as pd
```

```python
df = pd.read_csv("cityDataStuff.csv")
df = df.drop(columns = 'Headquarters')
df = df.drop(columns = 'Code')
df
```

```python
for i in df.columns:
    try:
        print(i)
        df[i].fillna(0, inplace=True)
        df[i] = pd.to_numeric(df[i])
        df[i] = df[i].astype(int)
    except:
        pass
df
```

```python
txtData = []
cities = df['District'].tolist()
df = df.drop(columns = 'District')
cities
```

```python
df.shape
```

```python
for i in df.columns:
    print(i)
    for j in range(df.shape[0]):
        txtData.append(str("The " + str(i) + " of " + str(cities[j]) + " District/City is " + str(df[i][j])))
```

```python
f = open("cityDataStuff.txt", 'w')
f.write('\n'.join(txtData))
f.close()
```
